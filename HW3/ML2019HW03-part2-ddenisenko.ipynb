````python
### BEGIN Solution

from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans, AgglomerativeClustering, Birch
from sklearn.metrics import silhouette_score, adjusted_mutual_info_score
from sklearn.model_selection import ParameterGrid
from matplotlib import pyplot as plt
%matplotlib inline

plt.figure(figsize=(20,15))
plt.tight_layout()
plt.rc('xtick', labelsize=7)   
plt.rc('ytick', labelsize=7)

models_labels = iter(['KMeans', 'GaussianMixture', 'Birch', 'Agglom'])

models = {
    KMeans: {
        'n_clusters': [1],
        'tol': [0.0001, 0.001, 0.01, 0.1], 
        'max_iter': [50, 100, 150, 200]
    }, 
    GaussianMixture: {
        'n_components': [1],
        'tol': [0.0001, 0.001, 0.01, 0.1], 
        'max_iter': [50, 100, 150, 200]
    }, 
    Birch: {
        'n_clusters': [1],
        'threshold': [0.1, 0.3, 0.5, 0.7], 
        'branching_factor': [30, 50, 80]
    }, 
    AgglomerativeClustering:{
        'n_clusters': [1],
        'linkage': ['ward', 'complete', 'average']
    }
}

n_clusters = [13,14,15,16,17]
plt_index = 1

for m_class, params in models.items():
    scores_by_cluster = {
        'silhouette': [],
        'ami': [],
    }
    mlabel = models_labels.__next__()
    for n_clust in n_clusters:
        estim = None
        best_silhouette = -1
        best_ami = -np.inf
        best_preds = None
        for g in ParameterGrid(params):
            if 'n_clusters' in g:
                g['n_clusters'] = n_clust
            if 'n_components' in g:
                g['n_components'] = n_clust
            
            m = m_class(**g)
            preds = m.fit_predict(points, labels)
            score_silhouette = silhouette_score(points, preds)
            score_ami = adjusted_mutual_info_score(labels, preds)
            
            if (score_silhouette >= best_silhouette) and (score_ami >= best_ami):
                estim = m
                best_silhouette = score_silhouette
                best_ami = score_ami
                best_preds = preds
        
        scores_by_cluster['silhouette'].append(best_silhouette)
        scores_by_cluster['ami'].append(best_ami)
        
        plt.subplot(len(models.keys()), len(n_clusters)+1, plt_index)
        plt_index += 1
        
        plt.scatter(points[:,0], points[:,1], c=best_preds, alpha=0.3, edgecolor='k')
        plt.title('{0}. N clusters: {1},\n silhouette = {2:.3f}, ami = {3:.3f}'.format(mlabel, n_clust, best_silhouette, best_ami))
        print(estim)
    
    plt.subplot(len(models.keys()), len(n_clusters)+1, plt_index)
    plt_index += 1
    
    plt.plot(n_clusters, scores_by_cluster['silhouette'])
    plt.plot(n_clusters, scores_by_cluster['ami'])
    
    plt.title('Silhouette and ami scores by number of clusters')
    plt.legend(['silhouette', 'ami'], loc='best')
    plt.xlabel('Number of clusters')
    plt.ylabel('Score')
        
plt.tight_layout()
plt.show()
````

# Dimentionality Reduction and Manifold Learning

## Task 2 (1 + 1 + 2 + 1 = 5 pt.)

Assume that you have $n$ points in $D$ dimensional space:

$$u_1, ..., u_n \in \mathbb R^D.$$

There exist a *linear* data transformation $F(u): \mathbb R^D \rightarrow \mathbb R^d, D >> d$ such that:

$$(1-\delta) \|u_i - u_j\|^2 \leq \|F(u_i) - F(u_j)\|^2\leq (1+\delta) \|u_i - u_j\|^2$$

with high probability.

The transformation $F(u)$ is: $F(u) = \tfrac{1}{\sqrt d}Xu$, and $X \in \mathbb R^{d \times D}$ is a random matrix, components of which are independent identically distributed $X_{ij} \sim \mathcal{N}(0, 1)$.

This statement means that there exists a matrix that reduces the original dimensionality such that pairwise distances are not distorted much. This is a theoretical probabilistic result that *guarantees* you such a transformation. We will obtain a bound on $d$ for which this result holds true.

**Chernoff inequality.** This unequality states a bound on distribution tail.

$$\mathbb P(X \geq t) \leq \frac{\mathbb E \exp(\lambda X)}{\exp(\lambda t)}$$

**Definition.** Random variable $X$ is called subexponential with parameters $(\sigma^2, b)$, if for all $\lambda: |\lambda| < \frac{1}{b}$ the following is true:

$$\mathbb E \exp(\lambda (X - \mathbb E X)) \leq \exp\left(\frac{\lambda^2\sigma^2}{2}\right)$$

**Fact.** $\chi^2$ distribution with $d$ degrees of freedom is a sum of squares of $d$ independent standard gaussian random variables.


**Fact.** $\chi^2$ distribution with $d$ degrees of freedom is subexponential with parameters $\sigma^2 = 4d, b = 4$

### Task 2.2 (1 pt.):

Prove that $\frac{\|Xu\|^2}{\|u\|^2}$ is $\chi_2$ random variable with $d$ degrees of freedom.

**BEGIN Solution**
Let's consider the elements of the vector $Xu$. It's easy to see that $i$th element is calculated via $\langle X_i, u \rangle$ where $\langle \cdot, \cdot \rangle$ is the scalar product of vectors.

Then,

$$\mathbb{E} \langle X_i, u \rangle/\|u\| = \langle \mathbb{E} X_i, u \rangle/\|u\| = 0$$
$$\mathbb{D} \langle X_i, u \rangle/\|u\| = \frac{1}{\|u\|^2}\mathbb{E}\langle X_i, u \rangle^2 = \frac{u^{\top}\mathbb{D}(X_i)u}{\|u\|^2} = \frac{u^{\top}\mathbb{I_d}u}{\|u\|^2} = 1$$


So $\langle X_i, u \rangle/\|u\| \sim {\cal N}(0, 1)$.

\begin{align}
\frac{\|Xu\|^2}{\|u\|^2} = \sum\limits_{i = 1}^d(X_i, u/\|u\|)^2 \sim \sum\limits_{i = 1}^d{\cal N}(0, 1) \overset{def.}{\sim} \chi^2(d)
\end{align}
**END Solution**

### Task 2.3 (2 pt.)

Having all of the previous results, we now may apply them to get the bound.

From inequality for tails from Task 4.1 we get that:

$$\mathbb P \left(\left\lvert \frac{\|Xu\|^2}{\|u\|^2} - d \right\rvert \geq t\right) \leq 2\exp\left(\frac{-t^2}{8d}\right)$$

This means that probability of such event that our distances will change a lot is bounded.

1. Show that if probability above is small, then probability of: $$(1-\delta) \|u_i - u_j\|^2 \leq \|\frac{1}{\sqrt d} X(u_i - u_j)\|^2\leq (1+\delta) \|u_i - u_j\|^2$$ is big and basically almost equal to $1 - n(n-1)\exp \left(-\frac{\delta^2d}{8}\right)$.

2. After the previous step, we got that if we want to have our inequalities to be true with high probability $1-\varepsilon$, we want to bound it from above. Derive the inequality for $d$ from that.

**Hint:**
* at some point you would like to take $\delta = \frac{t}{d}$. Note that it makes $\delta$ be in range of 0 and 1

**BEGIN Solution**

$$
\mathbb{P}\left(\left|\frac{\|Xu\|^2}{\|u\|^2} - d\right| \geq t\right) = \mathbb{P}\left(\frac{\|Xu\|^2}{\|u\|^2} \notin [d - t, d + t]\right) \overset{\delta = \frac{t}{d}}{=} \mathbb{P}\left(\frac{\|Xu\|^2}{\|\sqrt{d}u\|^2} \notin [1 - \delta, 1 + \delta]\right)
$$

$$
\mathbb{P}\left(\frac{\|Xu\|^2}{\|\sqrt{d}u\|^2} \notin [1 - \delta, 1 + \delta]\right) \leq 2\exp\left(-\frac{d\delta^2}{8}\right)
$$

It's also true that for $u_i \neq u_j$

$$
\mathbb{P}\left(\frac{\|X(u_i - u_j)\|^2}{\|\sqrt{d}(u_i - u_j)\|^2} \notin [1 - \delta, 1 + \delta]\right) \leq 2\exp\left(-\frac{d\delta^2}{8}\right)
$$

because $u_i - u_j \in \mathbb{R}^d$

Then

$$
\sum\limits_{u_i, u_j \in {\cal D}}\mathbb{P}\left(\frac{\|X(u_i - u_j)\|^2}{\|\sqrt{d}(u_i - u_j)\|^2} \in [1 - \delta, 1 + \delta]\right) = 1 - \sum\limits_{u_i, u_j \in {\cal D}}\mathbb{P}\left(\frac{\|X(u_i - u_j)\|^2}{\|\sqrt{d}(u_i - u_j)\|^2} \notin [1 - \delta, 1 + \delta]\right)
$$

And
$$
1 - \sum\limits_{u_i, u_j \in {\cal D}}\mathbb{P}\left(\frac{\|X(u_i - u_j)\|^2}{\|\sqrt{d}(u_i - u_j)\|^2} \notin [1 - \delta, 1 + \delta]\right) > 1 - n(n - 1)\exp\left(-\frac{d\delta^2}{8}\right)
$$

if $n(n - 1)\exp\left(-\frac{d\delta^2}{8}\right) \leq \varepsilon$, then

$$
1 - n(n - 1)\exp\left(-\frac{d\delta^2}{8}\right) \leq \varepsilon \geq 1-\varepsilon
$$

$$
n(n - 1)\exp\left(-\frac{d\delta^2}{8}\right) \leq \varepsilon
$$

$$
\exp \left(-\frac{d\delta^2}{8}\right) \leq \frac{\varepsilon}{n(n-1)}
$$

$$
\left(-\frac{d\delta^2}{8}\right) \leq \log \frac{\varepsilon}{n(n-1)}
$$

$$
\left(-\frac{t^2}{8d}\right) \leq \log \frac{\varepsilon}{n(n-1)}
$$

$$
d \geq -\frac{t^2}{8}\log \frac{\varepsilon}{n(n-1)}
$$
Which is equivalent to (if we hold $\delta$)
$$d \geq \frac{8}{\delta^2}\log\left(\frac{n(n-1)}{\varepsilon}\right)$$
**END Solution**

### Task 2.4 (1 pt.)

We have sucessfully got our lower bound on the dimensionality we can safely reduce to such that pairwise distances do not change much: 

$$d \geq \frac{8}{\delta^2}\log\left(\frac{n(n-1)}{\varepsilon}\right)$$

Note the beauty of that inequality. It **doesn't depend** on original dimensionality $D$, parameters $n$ -- number of samples and $\varepsilon$ -- probability are under the $\log$ function.

This bound is not very tight, using more advanced techniques you may improve it. That means, our estimate of $d$ may be too high, but we can guarantee our result for it. Also remember that this approach is probabilistic, and, basically, depends on how lucky you will be with your data. It is very useful in case of extremely large dimensionalities $D$, and allows to reduce dimensionality while preserving the structure of data.

````python
# basically I didn't realize how to perform it with PCA from sklearn, so I decided to use some code from our nla cource
# but acutally I realized it later, so, just scroll down
def compress(image, k):
    """
    Perform svd decomposition and truncate it (using k singular values/vectors)
    
    Parameters: 
        image (np.array):        input image (probably, colourful)
        
        k (int):                 approximation rank
        
    --------
    Returns:
      reconst_matrix (np.array): reconstructed matrix (tensor in colourful case)
      
      s (np.array):              array of singular values 
    """
#     Your code is here
    assert (image.shape[0] <= image.shape[1]) and (image.shape[0] <= image.shape[2]), "Reshape picture before passing it to this function"
    new_image = np.zeros(image.shape, dtype=np.float64)
    singular_values_for_each_channel = []
    for i in range(image.shape[0]):
        u,s,v = np.linalg.svd(image[i])
        u1 = u[:, :k]
        s1 = s[:k]
        v1 = v[:k, :]
        new_image[i] = u1.dot(np.diag(s1).dot(v1))
        singular_values_for_each_channel.append(s1)
    
    return new_image, singular_values_for_each_channel
````
